```python
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
import pandas as pd
```

```python
data = pd.read_csv("https://raw.githubusercontent.com/ezioauditore-tech/AI/main/datasets/Position_Salaries.csv")
data
```

```python
x = data.iloc[:,1:2]
print(x)
x = data.iloc[:,1:2].values
print(x)
```

```python
y = data.iloc[:,2].values
y
```

Linear Regression

```python
from sklearn.linear_model import LinearRegression
lin_reg = LinearRegression()
lin_reg.fit(x,y) #fit --> train
```

```python
plt.scatter(x,y,color = 'r')
plt.plot(x,lin_reg.predict(x),color = 'b')
plt.title("Linear Regression")
plt.xlabel("Position level")
plt.ylabel("Salary")
plt.show()
```

Polynomial Linear Regression with degree 2

```python
#degree rising
from sklearn.preprocessing import PolynomialFeatures #-->library
poly_reg = PolynomialFeatures(degree = 2)
x_poly = poly_reg.fit_transform(x)
```

```python
lin_reg2 = LinearRegression()
lin_reg2.fit(x_poly,y)
```

```python
plt.scatter(x,y,color = 'r')
plt.plot(x,lin_reg2.predict(x_poly),color = 'b')

plt.title("Polynomial Linear Regression with degree 2")
plt.xlabel("Position level")
plt.ylabel("Salary")
plt.show()
```

Polynomial Linear Regression with degree 3

```python
poly_reg = PolynomialFeatures(degree = 3)
x_poly3 = poly_reg.fit_transform(x)
lin_reg3 = LinearRegression()
lin_reg3.fit(x_poly3,y)
```

```python
plt.scatter(x,y,color = 'r')
plt.plot(x,lin_reg3.predict(x_poly3),color = 'b')

plt.title("Polynomial Linear Regression with degree 3")
plt.xlabel("Position level")
plt.ylabel("Salary")
plt.show()
```

```python
plt.scatter(x,y,color = 'r')
plt.plot(x,lin_reg2.predict(x_poly),color = 'g')
plt.plot(x,lin_reg3.predict(x_poly3),color = 'b')

plt.title("Polynomial Linear Regression with degree 2 & 3")
plt.xlabel("Position level")
plt.ylabel("Salary")
plt.show()
```

#Polynomial Regression from Scratch

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn import metrics
```

```python
data = pd.read_csv("https://raw.githubusercontent.com/ezioauditore-tech/AI/main/datasets/Real%20estate.csv ")
data
```

```python
data.head()
```

```python
data.shape
```

```python
data.info()
```

```python
data.corr()
```

```python
plt.subplots(figsize=(12,6))
sns.heatmap(data.corr(),annot = True,cmap = "Blues",)
```

```python
sns.displot(data = data,x = "Y house price of unit area",kde = True,color = "b")  #displot --> distribution plot
```

```python
x = data.drop("Y house price of unit area",axis=1)
x
```

```python
y = data["Y house price of unit area"]
y
```

taken polynomial features

```python
polynomial_reg = PolynomialFeatures(degree = 2 , include_bias = False)
poly_feat = polynomial_reg.fit_transform(x)
poly_feat.shape
```

split the test and train data

test size --> 30% (0.3)


```python
x_train,x_test,y_train,y_test = train_test_split(poly_feat,y,test_size = 0.3,random_state = 101)
```

Perform Linear Regression

and fit

```python
poly_model = LinearRegression()
poly_model.fit(x_train,y_train)
```

prediction

```python
y_pred = poly_model.predict(x_test)
```

testing and training

```python
pd.DataFrame({"y_test" : y_test , "y_pred" : y_pred , 'residuals' : (y_test - y_pred)})
```

error detection

```python
poly_MAE = metrics.mean_absolute_error(y_test,y_pred)
poly_MSE = metrics.mean_squared_error(y_test,y_pred) #mean squared error
poly_RMSE = np.sqrt(poly_MSE) # relative mean squared error
print(poly_MAE)
print(poly_MSE)
print(poly_RMSE)
```
