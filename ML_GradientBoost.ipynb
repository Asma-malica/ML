```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
test = pd.read_csv('https://raw.githubusercontent.com/ezioauditore-tech/AI/main/datasets/Titanic/test.csv')
train = pd.read_csv('https://raw.githubusercontent.com/ezioauditore-tech/AI/main/datasets/Titanic/train.csv')
```

```python
train.set_index('PassengerId',inplace = True)
test.set_index('PassengerId',inplace = True)
```

```python
train
```

```python
test
```

```python
y_train = train['Survived']
```

```python
train.drop(labels ='Survived',axis = 1 , inplace = True)
```

```python
train.shape
```

```python
train_test  = train.append(test) #as we need large dataset we are appending the test data along with the train data
train_test.shape # so now we got 1309 datas
train_test.isnull().sum()
```

```python
train_test.drop(columns=["Name","Ticket","Cabin","Age",'SibSp','Parch','Embarked'],axis=1,inplace=True)
```

```python
train_test
```

```python
train_test["Fare"].fillna(train_test["Fare"].mean(),inplace=True)
```

```python
train_test.isnull().sum()
```

```python
train_test.info()
```

```python

train_test_dummies = pd.get_dummies(train_test,columns={'Sex'})
```

```python
train_test_dummies.shape
```

```python
train_test_dummies.isnull().sum()
```

```python
train_test_dummies.fillna(value=0.0,inplace =True)
train_test_dummies.isnull().sum()
```

```python
x_train=train_test_dummies.values[0:891]
x_test=train_test_dummies.values[891:]
```

```python
from sklearn.preprocessing import MinMaxScaler
scalar = MinMaxScaler()
x_train_scale = scalar.fit_transform(x_train)
x_test_scale = scalar.transform(x_train)
```

```python
from sklearn.model_selection import train_test_split
x_train_sub,x_validation_sub,y_train_sub,y_validation_sub=train_test_split(x_train_scale,y_train,random_state=0)
```

```python
from sklearn.ensemble import GradientBoostingClassifier
from sklearn.metrics import classification_report,confusion_matrix
```

```python
learning_rates=[0.05,0.1,0.25,0.5,0.75,1]
for learning_rate in learning_rates:
    gb = GradientBoostingClassifier(n_estimators=20, learning_rate = learning_rate, max_features=2, max_depth = 2, random_state = 0)
    gb.fit(x_train_sub, y_train_sub)
    print("Learning rate: ", learning_rate)
    print("Accuracy score (training): {0:.3f}".format(gb.score(x_train_sub, y_train_sub)))
    print("Accuracy score (validation): {0:.3f}".format(gb.score(x_validation_sub, y_validation_sub)))
    print()
```
