#Decision Tree

* classify based on decisions

* can be number or categorical data

* supervised learning (labelled data)

* can be used to solve classification & regression

* uses tree representation to solve the problem
  
  1 . leaf node -> class label

  2 . Internal node -> Attributes

* paramters --> model a train panrathuku dataset larunthu  kodukra data

* hyperparamters --> learning model la tune panrathuku namma kodukura parameters

* Entropy --> impurities (randomness in data) ( 0 to 1) ( how much variance the data has )

* low entropy --> high information gain

* High entropy --> low information gain

* Entropy in inversely proportional to Information Gain

* Types Of DT

   1 . CART (Classification and Regression Tree) --> predictive model helps to find a variable based on other labeled variables (based on gini entrophy is calculated)
   
   2 . ID3 --> Iterative Dichotomiser 3

* Gini Impurity --> best fit identify panrathuku (Quality of Split)
   
   1 . G = sum(p(i) * 1-p(i))  --> (sum range 1 to c)

   Eg --> c = 2 , p(1)=p(2) = 0.7

   G = (0.7)(0.3) + (0.7)(0.3) --> 0.42

   G = 0.42

take i/p and o/p

```python
x = data.iloc[:,0:-1]
print(x)
y = data.iloc[:,-1]
print(y)
```

```python
from sklearn.model_selection import train_test_split # test_size --> 30% taken therefore for training 70%
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size = 0.3,random_state=0) #random state --> randomly picks the value in train data  #default value of random state is
```

```python
x.columns
```

train panra model

```python
from sklearn import tree,metrics
dtree = tree.DecisionTreeClassifier(criterion = "gini" ) #entrophy & gini are used to find the quality of the data
dtree.fit(x_train,y_train)
```

predicting

train

```python
y_train_pred = dtree.predict(x_train)
accuracy = metrics.accuracy_score(y_train,y_train_pred)  # train prediction score
print("Train Accuracy : {:.2f} ".format(accuracy))
print(" ")
accuracy = dtree.score(x_train,y_train)    # training score directly from model
print("Train Accuracy : {:.2f} ".format(accuracy))

```

test


```python
y_test_pred = dtree.predict(x_test)
```

```python
accuracy = metrics.accuracy_score(y_test,y_test_pred)
print("Test Accuracy : {:.2f} ".format(accuracy))
```
