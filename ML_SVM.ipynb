```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.svm import SVC
from sklearn import svm
```

Train data

```python
data = pd.read_csv("https://raw.githubusercontent.com/ezioauditore-tech/AI/main/datasets/Titanic/train.csv")
data
```

```python
data.drop(["PassengerId","Name","Ticket"],axis = 1,inplace = True)
data.head()
```

```python
data.isnull().sum()
```

```python
data.shape
```

```python
data[data.Embarked.isnull()]
```

converting the categorical data into numerical data (encoding)

```python
data.drop(["Cabin"] , axis = 1 ,inplace = True)
data["Age"] = data["Age"].fillna(data["Age"].median())
data["Embarked"] = data["Embarked"].fillna("C")
data.Sex = data.Sex.replace(["male","female"],[0,1])
data.Embarked = data.Embarked.replace(["S","C","Q"],[0,1,2])
```

```python
data.dropna(inplace = True)
data
```
