```python
import pandas as pd
import numpy as np
from sklearn.naive_bayes import GaussianNB
```

```python
data_train = pd.read_csv("https://raw.githubusercontent.com/nunnarilabs/ml/master/titanic/train.csv")
data_test = pd.read_csv("https://raw.githubusercontent.com/nunnarilabs/ml/master/titanic/test.csv")
passengerId = data_test["PassengerId"]
```

```python
final_data = data_train[["Survived","Pclass","Sex","Age","SibSp","Parch","Fare","Embarked"]]
final_data = final_data.dropna()
final_data.head()
```

```python
final_data.Sex = final_data["Sex"].replace(to_replace=final_data["Sex"].unique(),value=[1,0])
final_data
```

```python
final_data = pd.get_dummies(final_data,drop_first = True)
final_data
```

```python
train_y = final_data["Survived"]
train_x = final_data[["Pclass","Sex","Age","SibSp","Parch","Fare","Embarked_Q","Embarked_S"]]
```

```python
from sklearn.model_selection import train_test_split
data_train,final_data,train_target,val_target = train_test_split(train_x,train_y,train_size = 0.8)
```

```python
model = GaussianNB()
model.fit(data_train,train_target)
val_pred = model.predict(final_data)
```

```python
from sklearn.metrics import accuracy_score

print("model accuracy: {0:0.4f}",format(accuracy_score(val_target,val_pred)*100)+"%")
```
