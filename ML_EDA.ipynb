To Identify outliers

1.. Z score --> standard score --> how far the data is from the mean [(x-mean)/standard deviation] x --> value

2.. IQR( Inter Quantile Range ) ( 25% ,50% ,75% )

```python
from scipy import stats   #scipy for statistics (stats)
hp_zscore = np.abs(stats.zscore(data["normalized-losses"]))
hp_zscore
```

```python
threshold = hp_zscore.max() #outlier taken as threshold
index_val = np.where(hp_zscore>=threshold)
data.iloc[index_val]["normalized-losses"]
```

```python
new_data = data.drop(index=data.iloc[index_val]["normalized-losses"].index) #dropping the outlier
new_data
```

```python
q1 = data["length"].quantile(0.25)
q3 = data["length"].quantile(0.75)
IQR = q3 - q1
lower = q1 - 1.5 * IQR
upper  =  q3 + 1.5 * IQR
#create 2 arrays with boolean to store the outliers rows
upper_array = np.where(data['length'] >= upper)[0]
print(upper_array)  #outlier index (75% to 100%)
lower_array = np.where(data['length'] >= lower)[0]
print(lower_array)   #outlier index (0 to 25%)
```
