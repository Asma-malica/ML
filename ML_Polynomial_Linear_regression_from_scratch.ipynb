    * converting the polynomial / nonlinear  data to linear using degree (polynomial linear regression)


what happens if we increase the degree more higher ?? --> overfits the model (in training it will provide good result and testing it will provide bad result)

**Polynomial Regression from scratch**

steps for polynomial regression

     data --> polynomial --> linear --> prediction




```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.array([1,2,3,4,5])
y = np.array([2.1,3.8,6.5,9.2,11.3])
```

```python
degree = 2
```

```python
#design matrix --> column stack func

x_poly = np.column_stack([x**i for i in range(degree+1)])
x_poly
```

```python
weights = np.zeros(degree+1)
weights
```

```python
learning_rate = 0.001
num_iterations = 10#epochs
```

```python
for _ in range(num_iterations):
  #calculate predictions
  predictions = np.dot(x_poly,weights)
  print(weights)
  print(predictions)
  #calculate error
  error = predictions - y
  print(error)
  #update the weight using gradient descent
  gradient = np.dot(x_poly.T,error)/len(x)   #reduce the cost func
  weights -= learning_rate * gradient
  print(f"weights for iteration {_} : {weights}")
```

```python
learning_rate = 0.001
num_iterations = 10000 #epochs
for _ in range(num_iterations):
  #calculate predictions
  predictions = np.dot(x_poly,weights)
  # print(weights)
  # print(predictions)
  #calculate error
  error = predictions - y
  # print(error)
  #update the weight using gradient descent
  gradient = np.dot(x_poly.T,error)/len(x)   #reduce the cost func
  weights -= learning_rate * gradient
  print(f"weights for iteration {_} : {weights}")
```

```python
def predict(x,weights):
  x_poly = np.array([x**i for i in range(degree+1)])
  return np.dot(x_poly,weights)
```

```python
plt.scatter(x,y,label = "Data Points")
X_range = np.linspace(min(x),max(x),100)
Y_pred = [predict(x,weights) for x in X_range]
plt.plot(X_range,Y_pred,color = 'r' , label = "Polynomial Regression")
plt.xlabel('X')
plt.ylabel('Y')
plt.legend()
plt.show()
```
