```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def initialize_centroids(x,k):
  indices = np.random.choice(len(x),k,replace = False)
  return x[indices]
```

```python
def assign_to_cluster(x,centroids):
  distance=np.linalg.norm(x[:,np.newaxis]- centroids,axis = 2)
  return np.argmin(distance,axis=1)

```

```python
def update_centroids(x,labels,k):
  centroids=np.array([x[labels == i].mean(axis=0)for i in range(k)])
  return centroids
```

```python
def k_means(x,k,max_iters = 100):
  centroids = initialize_centroids(x,k)
  for _ in range(max_iters):
    labels = assign_to_cluster(x,centroids)
    new_centroids = update_centroids(x,labels,k)
    if(np.all(centroids==new_centroids)):
      break
    centroids = new_centroids
  return labels,centroids
```

```python
np.random.seed(42)  #seed = assign same random number to all
x=np.concatenate([np.random.normal(loc=i,scale=1,size=(50,2)) for i in range(5)])
print(x)
print(len(x))
```

```python
k=5
labels,centroids=k_means(x,k)
```

```python
plt.scatter(x[:, 0], x[:, 1], c=labels, cmap='viridis')
plt.scatter(centroids[:, 0], centroids[:, 1], marker='X', s=200, c='red', label='Centroids')
plt.title('K-means Clustering')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.legend()
plt.show()
```
